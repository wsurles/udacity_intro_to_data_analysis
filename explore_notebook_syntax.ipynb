### Timing code

````python
from math import sqrt

def fibol(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    return fibol(n-1) + fibol(n-2)

def fibol2(n):
    return ((1+sqrt(5))**n-(1-sqrt(5))**n)/(2**n*sqrt(5))
````

````python
fibol(20)
````

````python
%timeit fibol(20)
````

````python
fibol2(20)
````

````python
%timeit fibol2(20)
````
